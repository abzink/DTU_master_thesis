```python
import sys
sys.path.append('../')
import pandas as pd
from semi_supervised_graph import LabelPropagationAndSpreading
```

```python
df = pd.read_csv('/Users/annabzinkowska/DTU/master_thesis/data/processed_data_all_bertopic_vol_2.csv')
```

```python
categorical_columns = ['month_mode', 'quarter_mode','year_mode','day_week_mode', 'customer_country_mode']
numerical_columns = ['quantity_sum', 'price_sum', 'unit_weight', 'unit_price_mean', 'customer_country_count', 'customer_id_count'] 
text_column = 'description'
```

### Run Model

```python
# Example usage of the SemiSupervisedLearner class in another Jupyter notebook
learner = LabelPropagationAndSpreading(
    df=df,  # Your DataFrame
    categorical_cols=categorical_columns,  # List of categorical column names
    numerical_cols=numerical_columns,  # List of numerical column names
    target_col='category_bertopic',  # Name of the target column
    text_col=None,  # Name of the text column or None if not using text data
    use_text=False,  # Set to True if including text data
    output_file='predictions_run_4_20.csv',  # File to save the complete predictions
    eval_dir='saved_models_final',  # Directory to save models and individual prediction files
    verbose=True,
    labeled_percentage=0.2
)

# Perform the semi-spupervised learning and get the best models
best_models = learner.perform_learning()
```
