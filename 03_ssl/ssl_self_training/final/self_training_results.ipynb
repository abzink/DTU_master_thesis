```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'appended_df' is your DataFrame and it's already loaded in the environment.

# Filter out the data where SMOTE_Type is not None
filtered_df = appended_df[appended_df['SMOTE_Type'] != 'None']

# Define the new titles for each subplot
custom_titles = {
    'SMOTE': 'SMOTE',
    'BSMOTE': 'Borderline SMOTE',
    'No_SMOTE': 'No Balancing',
}

# Sort the unique SMOTE types based on the order in custom_titles
smote_types = sorted(filtered_df['SMOTE_Type'].unique(), key=lambda x: list(custom_titles.keys()).index(x))

# Set the desired bar width
bar_width = 0.16

# Create a figure with subplots - one for each SMOTE type (as ordered in custom_titles)
fig, axes = plt.subplots(1, len(smote_types), figsize=(20, 6), sharey=True, dpi=500)

# Use the second half of the default Seaborn color palette
palette = sns.color_palette("tab10")[5:]

# Loop through each unique SMOTE type (as ordered in custom_titles) and create a subplot for each
for i, smote_type in enumerate(smote_types):
    ax = axes[i]
    # Filter the DataFrame for the current SMOTE type
    data = filtered_df[filtered_df['SMOTE_Type'] == smote_type]
    
    # Plot the data for the current SMOTE type
    bars = sns.barplot(ax=ax, data=data, x='model', y='test_accuracy', hue='labeled_percentage', palette=palette)

         # Adjust bar width and find the highest bar
    max_height = 0
    max_x = 0
    for bar in bars.patches:
            bar.set_width(bar_width)
            if bar.get_height() > max_height:
                max_height = bar.get_height()
                max_x = bar.get_x() + bar.get_width() / 2

        # Mark the highest bar with a red star using ax.plot
        # Add a small value to max_height to position the star a bit higher
    star_offset = 0.04 * max_height  # Adjust this value as needed
    ax.plot(max_x, max_height + star_offset, "*", markersize=10, color="r")
    
    # Set custom title
    ax.set_title(custom_titles[smote_type], fontsize=14)
    
    # Rotate x-axis labels
    ax.set_xticklabels(ax.get_xticklabels(), rotation=0, fontsize=12)
    
    # Remove 'model' label
    ax.set_xlabel('')

# Set y-axis labels
axes[0].set_ylabel('Accuracy', fontsize=12)
for ax in axes[1:]:
    ax.set_ylabel('')

# Legend configuration - moved to the right side
handles, labels = axes[-1].get_legend_handles_labels()
fig.legend(handles, labels, title='Labeled Fraction (%)', loc='upper right', bbox_to_anchor=(0.63, 0.04), fontsize=12, ncol=5)

# Remove subplot legends
for ax in axes:
    ax.get_legend().remove()

# Set x-axis label for the figure
# fig.text(0.5, 0.02, 'Model', ha='center', fontsize=14)

# Set overall title
plt.suptitle('Self-Training Performance by Base Classifier', fontsize=16, y=0.95)

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show the plot
plt.show()
```
